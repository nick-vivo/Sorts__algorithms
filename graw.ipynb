```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

folder_path = 'data_base'
name_x = "Mass_Size"

for file_end in ["random.csv", "reserved.csv", "sort.csv"]:

    file_list = [f for f in sorted(os.listdir(folder_path)) if f.endswith(file_end)]

    for name_y in ["Time", "Copy_Count", "Comp_Count"]:
        plt.figure(figsize=(20, 15))  # Настроим размер окна
        axes = plt.axes()

        for file in file_list:

            file_path = os.path.join(folder_path, file)
            data = pd.read_csv(file_path, delimiter=',')
            x_train = np.array(data[name_x])
            y_train = np.array(data[name_y])
            fp = np.polyfit(x_train, y_train, 10)
            f = np.poly1d(fp)
            x_range = np.linspace(x_train.min(), x_train.max(), 5)

            plt.plot(x_range, f(x_range), label=file, linewidth=3)

        plt.title(f'График данных из файлов: ...{file_end}')
        plt.legend()
        # поменять параметр для лучшего изучения графиков
        axes.set_xlim([0, 5000])
        axes.set_ylim([0, 10000])

        axes.set_xlabel(name_x)
        axes.set_ylabel(name_y)
        plt.grid()
        plt.show()
```
